```python
#first random forest classifier using CCLE expression data
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import label_binarize
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import mutual_info_classif
from sklearn.multiclass import OneVsRestClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
encoder=LabelEncoder()
encoder.fit(y)
y=encoder.transform(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
classifier=mutual_info_classif(X_train,y_train)
n=600
indicies=np.argsort(classifier)[::-1][0:n]
```

```python
X_train=X_train[:,indicies]
X_test=X_test[:,indicies]
```

```python
# Use OneVsRest random forest because the performance was better 
#for huge numbers of features when researching random forest
clf = RandomForestClassifier(max_features =.2, bootstrap = True, random_state = 18)
forest=OneVsRestClassifier(clf)
forest.fit(X_train,y_train)
y_pred =forest.predict(X_test)
pred_prob = forest.predict_proba(X_test)
```

```python
from sklearn.metrics import balanced_accuracy_score,f1_score,precision_score, recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import roc_curve,auc
from sklearn.metrics import roc_auc_score
```

```python
#accuracy
accuracy=np.round(balanced_accuracy_score(y_test,y_pred),4)
print(f'accuracy:{accuracy}')

#precision
precision=np.round(precision_score(y_test,y_pred,average = 'weighted'),4)
print(f'precision:{precision}')

#recall
recall=np.round(recall_score(y_test,y_pred,average = 'weighted'),4)
print(f'recall:{recall}')

#f1score
f1score=np.round(f1_score(y_test,y_pred,average = 'weighted'),4)
print(f'f1score:{f1score}')
```
